#Agora se incia a análise de RFM
R = Recência
F = Frequência
M = Monetário


---


1. É necessário agregar os valores e aplicar alguns cálculos
2. Iremos analisar as compras dos clientes


```python
#Utilizaremos Grupby + agregação de cálculos
dados_agregado = tabela1.groupby('Customer ID').agg(
    max_data=('Order Date', 'max'),
    freq_vendas=('Order ID', 'count'),
    soma_vendas=('Sales', 'sum'),
).reset_index()

dados_agregado.head()
```

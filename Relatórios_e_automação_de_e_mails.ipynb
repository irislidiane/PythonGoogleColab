**Passo 1: importar base de dados**

```
import pandas as pd  #agora pandas se chama pd

#para ler o arquivo excel: 
tabela_vendas = pd.read_excel("/content/drive/MyDrive/Treinamentos/Hashtag Treinamentos - Programação/Aula 1 /Vendas.xlsx")

display(tabela_vendas)    #exibe o arquivo excel


```

**Passo 2: Calcular faturamento de cada loja:**

Filtrou-se as colunas necessáris entre colchetes duplo; agrupou-se com groupby. Soma-se o valor final de cada agrupamento com sum. Ordena-se com o sort_values, com o critério da coluna final em ordem decrescente.



```
tabela_faturamento = tabela_vendas[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
tabela_faturamento = tabela_faturamento.sort_values(by="Valor Final", ascending=False )
display(tabela_faturamento)
```

**Passo 3: Quantidade de produtos vendidos em cada loja**

Basicamente, este passo é parecido com o anterior, mudando o parâmetro.



```
tabela_quantidade = tabela_vendas[["ID Loja", "Quantidade"]].groupby("ID Loja").sum()
tabela_quantidade = tabela_quantidade.sort_values(by="Quantidade", ascending=False )
display(tabela_quantidade)
```

**Passo 4: Calcular o Ticket Médio dos produtos de cada loja**

Ticket médio = faturamento/quantidade

Ou seja, coluna "valor final"/ coluna de "quantidade
colocando os ***.to_frame()***, transformamos em tabela uma lista qualquer

```
ticket_medio = (tabela_faturamento["Valor Final"] / tabela_quantidade["Quantidade"]).to_frame()
ticket_medio = ticket_medio.rename(columns={0:"Ticket Médio"}) #renomeando a coluna 
display(ticket_medio)
```

**Passo 5: Enviar e-mail para diretoria**

**Passo 6: Enviar e-mail para cada loja**
 Enviando para uma unica loja.

```
lista_lojas = tabela_vendas["ID Loja"].unique() #mostrar a lista de lojas sem repetição
print(lista_lojas) 

loja = "Rio Mar Recife"
#tabela_vendas["ID Loja"] == "Rio Mar Recife" #comparando elementos na coluna ID Loja que são iguais a Rio Mar Recife para cada linha onde é verdadeiro
tabela_loja = tabela_vendas.loc[tabela_vendas["ID Loja"] == "Rio Mar Recife", ["ID Loja", "Quantidade", "Valor Final"]] #loc localiza linhas e colunas. loc[linas,colunas]
tabela_loja = tabela_loja.groupby("ID Loja").sum().join(ticket_medio)
display(tabela_loja)
```

**Enviando para cada loja**

```

for loja in lista_lojas:
  tabela_loja = tabela_vendas.loc[tabela_vendas["ID Loja"] == loja, ["ID Loja", "Quantidade", "Valor Final"]] #loc localiza linhas e colunas. loc[linas,colunas]
  tabela_loja = tabela_loja.groupby("ID Loja").sum().join(ticket_medio)
  enviar_email(loja, tabela_loja)
```
